```python
import os
import sys
import random
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

#from tqdm import tqdm
from itertools import chain
from skimage.io import imread, imshow, imread_collection, concatenate_images
from skimage.transform import resize
from skimage.morphology import label

from keras.models import Model, load_model
from keras.layers import Input, ZeroPadding2D
from keras.layers.core import Dropout, Lambda
from keras.layers.convolutional import Conv2D, Conv2DTranspose
from keras.layers.pooling import MaxPooling2D
from keras.layers.merge import concatenate
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras import backend as K
from keras import losses

import tensorflow as tf

warnings.filterwarnings('ignore', category=UserWarning, module='skimage')
seed = 42
random.seed = seed
np.random.seed = seed
tf.set_random_seed = seed
```

```python
# Set some parameters
IMG_WIDTH = 128
IMG_HEIGHT = 128
DROP_WIDTH = 32
DROP_HEIGHT = 32
IMG_CHANNELS = 1
TRAIN_PATH = '/media/oala/4TB/DATA/experiments-hhi/uncertainty-quant/ellipsoid-toy/f-only-rescaled-png-uint16/'
TEST_PATH = '/media/oala/4TB/DATA/experiments-hhi/uncertainty-quant/ellipsoid-toy/f-only-rescaled-png-uint16/'
RESIZE = True

CENTER_WIDTH = (IMG_WIDTH // 2) - (DROP_WIDTH // 2)
CENTER_HEIGHT = (IMG_HEIGHT// 2) - (DROP_HEIGHT // 2)
```

```python
# Get train and test IDs
train_names = next(os.walk(TRAIN_PATH))[2]
test_names = next(os.walk(TEST_PATH))[2]
```

```python
# Get and resize train images and masks
X_train = np.zeros((len(train_names), IMG_HEIGHT, IMG_WIDTH), dtype=np.uint16)
Y_train = np.zeros((len(train_names), DROP_HEIGHT, DROP_WIDTH), dtype=np.float64)
print('Getting and resizing train images and masks ... ')
sys.stdout.flush()
for image_name, count in zip(train_names, range(len(train_names))):
    img = imread(TRAIN_PATH + image_name)
    img = resize(img, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True)
    X_train[count] = img
    #take out center
    X_train[count,CENTER_HEIGHT:CENTER_HEIGHT+DROP_HEIGHT,CENTER_WIDTH:CENTER_WIDTH+DROP_WIDTH] = 0
    #use center as reference
    Y_train[count] = img[CENTER_HEIGHT:CENTER_HEIGHT+DROP_HEIGHT,CENTER_WIDTH:CENTER_WIDTH+DROP_WIDTH]
    if count % 500 == 0:
        print('Done with # ', count)

# Get and resize test images
X_test = np.zeros((len(test_names), IMG_HEIGHT, IMG_WIDTH), dtype=np.uint16)
Y_test = np.zeros((len(test_names), DROP_HEIGHT, DROP_WIDTH), dtype=np.float64)
print('Getting and resizing train images and masks ... ')
sys.stdout.flush()
for image_name, count in zip(train_names, range(len(train_names))):
    img = imread(TEST_PATH + image_name)
    img = resize(img, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True)
    X_test[count] = img
    #take out center
    X_test[count,CENTER_HEIGHT:CENTER_HEIGHT+DROP_HEIGHT,CENTER_WIDTH:CENTER_WIDTH+DROP_WIDTH] = 0
    #use center as reference
    Y_test[count] = img[CENTER_HEIGHT:CENTER_HEIGHT+DROP_HEIGHT,CENTER_WIDTH:CENTER_WIDTH+DROP_WIDTH]
    if count % 500 == 0:
        print('Done with # ', count)
#add channel axis to data
X_train = X_train[:,:,:,np.newaxis]
Y_train = Y_train[:,:,:,np.newaxis]
X_test = X_test[:,:,:,np.newaxis]
Y_test = Y_test[:,:,:,np.newaxis]

print('Done!')
```

```python
# Build U-Net model
inputs = Input((IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS))
s = Lambda(lambda x: x / 65535) (inputs)

c1 = Conv2D(64, (4, 4), strides=(2, 2), activation='elu', kernel_initializer='he_normal', padding='valid') (s)
c1 = ZeroPadding2D(((1,0),(1,0))) (c1)
#c1 = Dropout(0.1) (c1)
#c1 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c1)
#p1 = MaxPooling2D((2, 2)) (c1)

c2 = Conv2D(64, (4, 4), strides=(2, 2), activation='elu', kernel_initializer='he_normal', padding='valid') (c1)
#c2 = Dropout(0.1) (c2)
#c2 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c2)
#p2 = MaxPooling2D((2, 2)) (c2)
c2 = ZeroPadding2D(((1,0),(1,0))) (c2)

c3 = Conv2D(128, (4, 4), strides=(2, 2), activation='elu', kernel_initializer='he_normal', padding='valid') (c2)
#c3 = Dropout(0.2) (c3)
#c3 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c3)
#p3 = MaxPooling2D((2, 2)) (c3)
c3 = ZeroPadding2D(((1,0),(1,0))) (c3)

c4 = Conv2D(256, (4, 4), strides=(2, 2), activation='elu', kernel_initializer='he_normal', padding='valid') (c3)
#c4 = Dropout(0.2) (c4)
#c4 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c4)
#p4 = MaxPooling2D(pool_size=(2, 2)) (c4)
c4 = ZeroPadding2D(((1,0),(1,0))) (c4)

c5 = Conv2D(512, (4, 4), strides=(2, 2), activation='elu', kernel_initializer='he_normal', padding='valid') (c4)
#c5 = Dropout(0.3) (c5)
#c5 = Conv2D(256, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same') (c5)
c5 = ZeroPadding2D(((1,0),(1,0))) (c5)

c6 = Conv2D(4000, (4, 4), strides=(2, 2), activation='elu', kernel_initializer='he_normal', padding='valid') (c5)

uc1 = Conv2DTranspose(512, (4, 4), activation='relu', kernel_initializer='he_normal', padding='valid') (c6)

uc2 = Conv2DTranspose(256, (4, 4), strides=(2, 2), activation='relu', kernel_initializer='he_normal') (uc1)
uc2 = ZeroPadding2D(((1,0),(1,0))) (uc2)

uc3 = Conv2DTranspose(128, (4, 4), strides=(2, 2), activation='relu', kernel_initializer='he_normal', padding='valid') (uc2)
uc3 = ZeroPadding2D(((1,0),(1,0))) (uc3)

uc4 = Conv2DTranspose(64, (4, 4), strides=(2, 2), activation='relu', kernel_initializer='he_normal', padding='valid') (uc3)
uc4 = ZeroPadding2D(((1,0),(1,0))) (uc4)

outputs = Conv2DTranspose(1, (1, 1), strides=(2, 2), activation='sigmoid', kernel_initializer='he_normal', padding='valid') (uc4)

model = Model(inputs=[inputs], outputs=[outputs])
model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```
