```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

import scipy.io
import re
import os
import numpy as np
import pickle

from matplotlib.lines import Line2D

def tryint(s):
    try:
        return int(s)
    except ValueError:
        return s

def alphanum_key(s):
    """ Turn a string into a list of string and number chunks.
        "z23a" -> ["z", 23, "a"]
    """
    return [ tryint(c) for c in re.split('([0-9]+)', s) ]

def get_data(path):
    file_names = next(os.walk(path))[2]
    file_names.sort(key=alphanum_key)
    X = []
    Y = []
    print('Reading data ... ')
    for file_name, count in zip(file_names, range(len(file_names))):
        mat = scipy.io.loadmat(path + file_name)
        X.append(mat['y'])
        Y.append(mat['x'])
        if count % 500 == 0:
            print('Done with # ', count)
    X = np.array(X)
    Y = np.array(Y)
    
    print(X.shape)
    print(Y.shape)
    
    return X[:1600], Y[:1600], X[1600:1800], Y[1600:1800], X[1800:], Y[1800:]

    
def get_visualization_data(path_to_epoch_dicts_folder_testdata, path_to_dataset, error_type):
    '''
    Input:
    path_to_epoch_dicts_folder_testdata (string): string that points to folder containing the dict with preds and proxy scores of different epochs
    path_to_dataset (string): string that points to folder containinf the entire dataset with inputs and targets
    error_type (string): specifies the type of error
    
    Returns:
    target_signal (np.array): has dims [num data points, d]
    predicted_signal (np.array): has dims [num epochs, num data doints, d]
    error (np.array): has dims [num epochs, num data doints, d]
    proxy (np.array): has dims [num epochs, num data doints, d]
    '''
    
    file_names = next(os.walk(path_to_epoch_dicts_folder_testdata))[2]
    file_names.sort(key=alphanum_key)
    
    predicted_signal = []
    proxy = []
    
    for file_name, count in zip(file_names, range(len(file_names))):
        matrices_path = path_to_epoch_dicts_folder_testdata+file_name
        with open(matrices_path, 'rb') as handle:
            matrices = pickle.load(handle)
        predicted_signal.append(matrices['pred'])
        proxy.append(matrices['uncertainty'])
    
    
    predicted_signal = np.array(predicted_signal)
    proxy = np.array(proxy)
    _, _, _, _, _, target_signal = get_data(path_to_dataset)
    
    
    if len(predicted_signal.shape) < 4:
        predicted_signal = predicted_signal[:,:,:,np.newaxis]
    
    if error_type == 'abs':
        error = np.abs(target_signal - predicted_signal)
    elif error_type == 'squared_diff':
        error = (target_signal - predicted_signal)**2
    elif error_type == 'diff':
        error = target_signal - predicted_signal
    else:
        print('Problem when getting the data for visualization: No valid error was specified')
        
    return target_signal, predicted_signal, error, proxy

def make_gif(target_signal, predicted_signal, error, proxy, 
             error_type, proxy_type, save_filename, dpi, interval, 
             point_number, train_data_dist, test_data_dist,
            error_ylim, proxy_ylim):
    '''
    Input:
    #fixed values
    target_signal (np.array): should have dimension [d]

    #values being updated
    predicted_signal (np.array): should have dimension [num_epochs, d]
    error (np.array): should have dimension [num_epochs, d]
    proxy (np.array): should have dimension [num_epochs, d]
    
    #misc variables
    error_type (string): specifies the type of error (for labeling)
    proxy_type (string): specifies the type of proxy (for labeling)
    save_filename (string): with ending .gif
    dpi (int): resolution of gif framees
    interval (int): time in ms between frames
    point_number (int): id of the test data point we are tracking
    train_data_dist (string): name of train data dist
    test_data_dist (string): name of test data dist
    error_ylim (tuple): (min, max)
    proxy_ylim (tuple): (min, max)
    
    Returns:
    None

    '''
    plt.rcParams.update({'font.size': 14})
    
    fig, ((ax1), (ax2)) = plt.subplots(nrows = 2, ncols = 1, figsize=(20,6))
    #fig.set_tight_layout(True)
    #fig.tight_layout()
    fig.subplots_adjust(top=0.8, hspace=0.5)

    # Query the figure's on-screen size and DPI. Note that when saving the figure to
    # a file, we need to provide a DPI for that separately.
    print('fig size: {0} DPI, size in inches {1}'.format(
        fig.get_dpi(), fig.get_size_inches()))

    # Plot a scatter that persists (isn't redrawn) and the initial line.
    d = np.arange(target_signal.shape[0])
    epochs = error.shape[0]

    #manage axis design
    ax2.tick_params('y', colors='red')
    ax3 = ax2.twinx()
    ax3.tick_params('y', colors='black')
    ax2.set_ylim(error_ylim)
    #ax3.set_ylim([0,0.0017])
    ax3.set_ylim(proxy_ylim)
    ax2.set_ylabel(error_type+' error', color = 'red')
    ax3.set_ylabel(proxy_type+'  error proxy', color = 'black')
    
    #all thats fixed
    ax1.plot(d, target_signal, label = 'target signal', color = 'green', lw=2.5)

    #all that changes
    predicted_signal_line, = ax1.plot(d, predicted_signal[0], color = 'deepskyblue', label = 'predicted signal', lw=2.5)
    error_line, = ax2.plot(d, error[0], color = 'red', label = error_type+' error', lw=2.5)
    proxy_line, = ax3.plot(d, error[0], color = 'black', label = proxy_type+' error proxy', lw=2.5)
    
    #ax1.legend(loc=1)
    #ax2.legend(loc=1)
    #ax3.legend(loc=5)
    #ax1.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    #ax2.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    #ax3.legend(bbox_to_anchor=(1.05, 1), loc=3, borderaxespad=0.)
    ax1.legend(bbox_to_anchor=(0., -0.4, 1., .102), loc=3,
           ncol=2, borderaxespad=0.)
    ax2.legend(bbox_to_anchor=(0., -0.4, 1., .102), loc=3,
           ncol=1, borderaxespad=0.)
    ax3.legend(bbox_to_anchor=(0., -0.4, 1., .102), loc=4,
           ncol=1, borderaxespad=0.)

    def update(i):
        #label = 'Epoch {0}'.format(i) + ', test data point #' + str(point_number)
        print(label)
        # Update the line and the axes (with a new xlabel). Return a tuple of
        # "artists" that have to be redrawn for this frame.

        predicted_signal_line.set_ydata(predicted_signal[i])
        error_line.set_ydata(error[i])
        proxy_line.set_ydata(proxy[i])

        #ax2.set_xlabel(label)
        #ax1.set_title(label, fontsize=16)
        #fig.suptitle(label, fontsize=16)
        fig.suptitle('Error proxy method: {} - Error type: {} - Train data: {} - Test data: {} - Test point ID: {} \n\n Epoch: {}'.format(proxy_type, 
                                                         error_type, 
                                                         train_data_dist, 
                                                         test_data_dist, 
                                                         point_number,
                                                          i
                                                        ), fontsize=16)
        return predicted_signal_line, error_line, proxy_line, fig #, ax2


    # FuncAnimation will call the 'update' function for each frame; here
    # animating over 10 frames, with an interval of 200ms between frames.
    anim = FuncAnimation(fig, update, frames=np.arange(0, epochs), interval=interval)
    anim.save(save_filename, dpi=dpi, writer='imagemagick')
    
    
def make_gif_4(target_signal, predicted_signals, errors, proxys, 
             error_type, proxy_types, save_filename, dpi, interval, 
             point_number, train_data_dist, test_data_dist,
            error_ylims, proxy_ylims, epochs):
    '''
    Input:
    #fixed values
    target_signal (np.array): should have dimension [num_datapoints, d]

    #values being updated
    predicted_signals (list of 4 np.arrays): each should have dimension [num_epochs,num_datapoints, d]
    errors (list of 4 np.arrays): each should have dimension [num_epochs,num_datapoints, d]
    proxys (list of 4 np.arrays): each should have dimension [num_epochs,num_datapoints, d]
    
    #misc variables
    error_type (string): specifies the type of error (for labeling)
    proxy_types (list of 4 strings): specifies the type of proxy (for labeling)
    save_filename (string): with ending .gif
    dpi (int): resolution of gif framees
    interval (int): time in ms between frames
    point_number (int): id of the test data point we are tracking
    train_data_dist (string): name of train data dist
    test_data_dist (string): name of test data dist
    error_ylim (list of 4 tuples): each has (min, max)
    proxy_ylims (list of 4 tuples): each has (min, max)
    epochs (np.array): contains the ids of epochs displayed
    
    Returns:
    None

    '''
    plt.rcParams.update({'font.size': 14})
    
    fig, ((ax1, ax3), (ax2, ax4), (ax5, ax7), (ax6, ax8)) = plt.subplots(nrows = 4, ncols = 2, figsize=(30,14))
    #fig.set_tight_layout(True)
    #fig.tight_layout()
    fig.subplots_adjust(top=0.9, hspace=0.5)

    # Query the figure's on-screen size and DPI. Note that when saving the figure to
    # a file, we need to provide a DPI for that separately.
    print('fig size: {0} DPI, size in inches {1}'.format(
        fig.get_dpi(), fig.get_size_inches()))

    # Plot a scatter that persists (isn't redrawn) and the initial line.
    d = np.arange(target_signal.shape[1])

    #manage axis design
    ax1.set_title('Error proxy method: '+ proxy_types[0])
    ax3.set_title('Error proxy method: '+ proxy_types[1])
    ax5.set_title('Error proxy method: '+ proxy_types[2])
    ax7.set_title('Error proxy method: '+ proxy_types[3])
    
    ax2.tick_params('y', colors='red')
    ax4.tick_params('y', colors='red')
    ax6.tick_params('y', colors='red')
    ax8.tick_params('y', colors='red')
    
    ax2.set_ylim(error_ylims[0])
    ax2.set_ylabel(error_type+' error', color = 'red')
    ax4.set_ylim(error_ylims[1])
    ax4.set_ylabel(error_type+' error', color = 'red')
    ax6.set_ylim(error_ylims[2])
    ax6.set_ylabel(error_type+' error', color = 'red')
    ax8.set_ylim(error_ylims[3])
    ax8.set_ylabel(error_type+' error', color = 'red')
    
    #make twin axes
    ax2_1 = ax2.twinx()
    ax2_1.tick_params('y', colors='black')
    ax4_1 = ax4.twinx()
    ax4_1.tick_params('y', colors='black')
    ax6_1 = ax6.twinx()
    ax6_1.tick_params('y', colors='black')
    ax8_1 = ax8.twinx()
    ax8_1.tick_params('y', colors='black')
    
    
    #manage twin axis design
    ax2_1.set_ylim(proxy_ylims[0])
    ax2_1.set_ylabel('error proxy', color = 'black')
    ax4_1.set_ylim(proxy_ylims[1])
    ax4_1.set_ylabel('error proxy', color = 'black')
    ax6_1.set_ylim(proxy_ylims[2])
    ax6_1.set_ylabel('error proxy', color = 'black')
    ax8_1.set_ylim(proxy_ylims[3])
    ax8_1.set_ylabel('error proxy', color = 'black')
    
    #all thats fixed
    ax1.plot(d, target_signal[point_number], label = 'target signal', color = 'green', lw=2.5)
    ax3.plot(d, target_signal[point_number], label = 'target signal', color = 'green', lw=2.5)
    ax5.plot(d, target_signal[point_number], label = 'target signal', color = 'green', lw=2.5)
    ax7.plot(d, target_signal[point_number], label = 'target signal', color = 'green', lw=2.5)

    #all that changes
    predicted_signal_line_1, = ax1.plot(d, predicted_signals[0][0, point_number], color = 'deepskyblue', label = 'predicted signal', lw=2.5)
    error_line_1, = ax2.plot(d, errors[0][0, point_number], color = 'red', label = error_type+' error', lw=2.5)
    proxy_line_1, = ax2_1.plot(d, proxys[0][0, point_number], color = 'black', label = proxy_types[0]+' error proxy', lw=2.5)
    
    predicted_signal_line_2, = ax3.plot(d, predicted_signals[1][0, point_number], color = 'deepskyblue', label = 'predicted signal', lw=2.5)
    error_line_2, = ax4.plot(d, errors[1][0, point_number], color = 'red', label = error_type+' error', lw=2.5)
    proxy_line_2, = ax4_1.plot(d, proxys[1][0, point_number], color = 'black', label = proxy_types[1]+' error proxy', lw=2.5)
    
    predicted_signal_line_3, = ax5.plot(d, predicted_signals[2][0, point_number], color = 'deepskyblue', label = 'predicted signal', lw=2.5)
    error_line_3, = ax6.plot(d, errors[2][0, point_number], color = 'red', label = error_type+' error', lw=2.5)
    proxy_line_3, = ax6_1.plot(d, proxys[2][0, point_number], color = 'black', label = proxy_types[2]+' error proxy', lw=2.5)
    
    predicted_signal_line_4, = ax7.plot(d, predicted_signals[3][0, point_number], color = 'deepskyblue', label = 'predicted signal', lw=2.5)
    error_line_4, = ax8.plot(d, errors[3][0, point_number], color = 'red', label = error_type+' error', lw=2.5)
    proxy_line_4, = ax8_1.plot(d, proxys[3][0, point_number], color = 'black', label = proxy_types[3]+' error proxy', lw=2.5)
    
    #TODO: manage legends 
    '''
    ax1.legend(bbox_to_anchor=(0., -0.4, 1., .102), loc=3,
           ncol=2, borderaxespad=0.)
    ax2.legend(bbox_to_anchor=(0., -0.4, 1., .102), loc=3,
           ncol=1, borderaxespad=0.)
    ax2_1.legend(bbox_to_anchor=(0., -0.4, 1., .102), loc=4,
           ncol=1, borderaxespad=0.)
    '''
    custom_lines = [Line2D([0], [0], color='green', lw=4),
                Line2D([0], [0], color='deepskyblue', lw=4),
                Line2D([0], [0], color='red', lw=4),
                   Line2D([0], [0], color='black', lw=4)]
    
    captions = ['target signal', 'predicted signal', error_type+' error', 'error proxy']
    
    ax6.legend(custom_lines, captions, bbox_to_anchor=(0.5, -0.5, 1.25, .102), loc=4,
           mode='expand', ncol=4, borderaxespad=0.)

    
    
    def update(i):
        #label = 'Epoch {0}'.format(i) + ', test data point #' + str(point_number)
        print(label)
        # Update the line and the axes (with a new xlabel). Return a tuple of
        # "artists" that have to be redrawn for this frame.

        predicted_signal_line_1.set_ydata(predicted_signals[0][i,point_number])
        error_line_1.set_ydata(errors[0][i,point_number])
        proxy_line_1.set_ydata(proxys[0][i,point_number])
        
        predicted_signal_line_2.set_ydata(predicted_signals[1][i, point_number])
        error_line_2.set_ydata(errors[1][i,point_number])
        proxy_line_2.set_ydata(proxys[1][i,point_number])
        
        predicted_signal_line_3.set_ydata(predicted_signals[2][i,point_number])
        error_line_3.set_ydata(errors[2][i,point_number])
        proxy_line_3.set_ydata(proxys[2][i,point_number])
        
        predicted_signal_line_4.set_ydata(predicted_signals[3][i, point_number])
        error_line_4.set_ydata(errors[3][i,point_number])
        proxy_line_4.set_ydata(proxys[3][i,point_number])

        #ax2.set_xlabel(label)
        #ax1.set_title(label, fontsize=16)
        #fig.suptitle(label, fontsize=16)
        fig.suptitle('Error type: {} - Train data: {} - Test data: {} - Test point ID: {} \n\n Epoch: {}'.format(error_type, 
                                                         train_data_dist, 
                                                         test_data_dist, 
                                                         point_number,
                                                          epochs[i]+1
                                                        ), fontsize=16)
        return predicted_signal_line_1, error_line_1, proxy_line_1, predicted_signal_line_2, error_line_2, proxy_line_2, predicted_signal_line_3, error_line_3, proxy_line_3, predicted_signal_line_4, error_line_4, proxy_line_4, fig #, ax2


    # FuncAnimation will call the 'update' function for each frame; here
    # animating over 10 frames, with an interval of 200ms between frames.
    anim = FuncAnimation(fig, update, frames=np.arange(0, len(epochs)), interval=interval)
    anim.save(save_filename, dpi=dpi, writer='imagemagick')
```

```python
import scipy

#scipy.spatial.distance.cosine(u, v)

def mse_calc(proxy,error):
    
    ses = []
    for i in range(proxy.shape[0]):
        ses.append(((error[i]-proxy[i])**2).sum())
    
    ses = np.array(ses)
    #return((error-proxy)**2).sum()/proxy.shape[0]
    #print(ses.shape)
    return np.mean(ses)

def cos_calc(proxy,error):
    '''
    top = (proxy*error).sum(axis=1)
    bottom1 = (proxy**2).sum(axis=1)
    bottom2 = (proxy**2).sum(axis=1)
    
    print(top.shape, bottom1.shape, bottom2.shape)
    return (top/(bottom1*bottom2)).sum()#/proxy.shape[0]
    '''
    cosines = []
    for i in range(proxy.shape[0]):
        cosines.append(scipy.spatial.distance.cosine(proxy[i], error[i]))
    cosines = np.array(cosines)
    #print(cosines)
    return np.mean(cosines)

def compute_metrics(proxys, errors, selected_epochs):
    for proxy, error, proxy_type in zip(proxys, errors, ['drop', 'error_pred', 'interval', 'prob']):
        print('\n')
        for epoch, epoch_id in zip(range(proxy.shape[0]), selected_epochs):
            mse = mse_calc(proxy[epoch], error[epoch])
            cos = cos_calc(proxy[epoch], error[epoch])
            print('{} at epoch {} - mse: {} cos: {}'.format(proxy_type, epoch_id, mse, cos))
```
